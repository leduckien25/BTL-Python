```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Firefox()
origin_url = "https://www.footballtransfers.com/us/transfers/confirmed/most-recent/2024-2025/uk-premier-league/"

transfers = {}

for i in range(1, 17):
    current_url  = origin_url + str(i)

    driver.get(current_url)
    time.sleep(3)

    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')

    tbody = soup.find('tbody', id='player-table-body')

    cells = tbody.find_all('tr')


    for tr in cells:
        player_name = tr.find('td', class_='td-player').span.get_text().strip()

        transfer_value = tr.find('td', class_='td-price').span.get_text().strip()

        transfers[player_name] = transfer_value
        
driver.quit()
transfers
        
```

```python
import sqlite3

conn = sqlite3.connect('players.db')
cursor = conn.cursor()

cursor.execute('SELECT player FROM players')
players_in_db = [row[0] for row in cursor.fetchall()]
```

```python
print(len(players_in_db))

player_in_db_transfers = []

for player in players_in_db:
    if player not in transfers:
        player_in_db_transfers.append({'player': player, 'value': 'N/A'})
    else:
        player_in_db_transfers.append({'player':player, 'value': transfers[player]})

player_in_db_transfers
```

```python
import sqlite3

import pandas as pd

conn = sqlite3.connect('players.db')
cursor = conn.cursor()
cursor.execute("DROP TABLE IF EXISTS player_transfer_values")
cursor.execute('CREATE TABLE player_transfer_values(player text, value text)')
conn.commit()

df = pd.DataFrame(player_in_db_transfers)

df.to_sql("player_transfer_values", conn, if_exists="append", index=False)

conn.close()
```
